```python
# dfimg = pd.DataFrame("C:/Users/a21ma/OneDrive/Desktop/Code/Machine Learning/Weapon Detection/Dataset/weapon_detection/train/images")
import cv2
import os

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images

samples = load_images_from_folder("C:/Users/a21ma/OneDrive/Desktop/Code/Machine Learning/Weapon Detection/Dataset/weapon_detection/train/images")
```

```python
datagen1 = ImageDataGenerator(width_shift_range=[-100,100])

datagen2 = ImageDataGenerator(height_shift_range=[-100,100])

datagen3 = ImageDataGenerator(horizontal_flip=True)

datagen4 = ImageDataGenerator(vertical_flip=True)

datagen5 = ImageDataGenerator(rotation_range=50)

datagen6 = ImageDataGenerator(brightness_range=[0.2,1.0])

datagen7 = ImageDataGenerator(zoom_range=[0.5,1.0])
```

```python
data=[]
```

```python
# Assuming 'samples' is a list of input images (numpy arrays)

for sample in samples:
    sample = expand_dims(sample, 0)
    it2 = datagen1.flow(sample, batch_size=1)
    for i in range(3):
        # plt.subplot(330 + 1 + i)
        # generate batch of images
        batch = it2.next()
        # convert to unsigned integers for viewing
        image = batch[0].astype('uint8')
        # plot raw pixel data
        # plt.imshow(image)
        data.append(image)

    # show the figure after generating each set of augmented images
    # plt.show()
```
